В ряде случаев вместо рейтинга, выраженного от 1 до 10 стоят числа с процентами. Их нужно преобразовать.

```python
#c помощью специального метода уберем % и заменим тип данных на float
data['ratings'].str.rstrip("%").astype(float).unique()
```

```python
data['ratings'] = data['ratings'].str.rstrip("%").astype(float)
```

Мы убрали %, но теперь часть данных в стобалльной системе. Поделим такие значения на 10, чтобы привести все данные в десятибальную систему.

```python
#напишем для этого простую функцию, которая будет проверять, больше ли значение 10, и если да, то делить на 10
def division_ten(rating):
    if rating > 10:
        rating = rating / 10
    return rating
```

```python
#применяем функцию
data['ratings'] = data['ratings'].apply(division_ten)
```

```python
#смотрим, все ли значения корректно перевелись
data['ratings'].unique()
```

Последним этапом является анализ того, сколько фильмов принесли хоть что-то создателям. Для этого создадим поле 'paid_of' (окупились). В среднем, чтобы фильм окупился, он должен собрать в два раза больше, чем на него было потрачено, так как примерно половину доходов со сборов забирают кинотеатре. В новом поле будет True, если сборы в 2 и больше раз превосходят бюджет и False, если нет.

```python
data_2015_2019_support['paid_of'] = data_2015_2019_support['box_office'] >= data_2015_2019_support['budget'] * 2
```

```python
#построим гистограму, в которой будет указано, какая доля фильмов с гос.поддержкой окупается
data_2015_2019_support.pivot_table(index='year', values='paid_of', aggfunc='mean').plot(kind='bar')
plt.ticklabel_format(style='plain', axis='y')
```
